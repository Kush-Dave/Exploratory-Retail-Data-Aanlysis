STEP-1 IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import plotly.express as pe
import warnings
warnings.filterwarnings('ignore')
print("The libraries are imported 100%")
```

STEP-2 IMPORTING DATASET

```python
df =  pd.read_csv("https://raw.githubusercontent.com/Kush-Dave/Exploratory-Retail-Data-Aanlysis/main/SampleSuperstore.csv")
print("The dataset has been imported 100%")
```

STEP-4 CORRELATION MATRIX

```python
corr = df.corr()
sb.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, 
            cmap=sb.diverging_palette(220, 20, as_cmap=True))
```

```python
sb.catplot(x="Segment",y="Profit",data=df,kind='bar',ci=90)
```

```python
sb.catplot(x="Ship Mode",y="Profit",data=df,kind='bar',ci=None)
```

```python
order = df.groupby(["State"])["Profit"].mean().sort_values().index
sb.catplot(y="State",x="Profit",data=df,kind='bar',ci=None,height=20,order=order)
```

```python
sb.catplot(x="Category",y="Profit",data=df,kind="bar",ci=None)
```

```python
sb.catplot(x="Category",y="Profit",data=df,kind='bar',ci=None,col="Region")
```

```python
plt.subplots(figsize=(8,10))
order=df.groupby(['State'])["State"].count().sort_values().index
sb.countplot(y='State',data=df,order=order)
```
